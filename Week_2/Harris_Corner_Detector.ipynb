```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
filename = '/gdrive/MyDrive/Colab Notebooks/Computer Vision/Week 2/ChessBoard.png'
img = cv2.imread(filename)
cv2_imshow(img)
```

- **img - Input image**. It should be grayscale and float32 type.
- **blockSize** - It is the size of neighbourhood considered for corner detection
- **ksize** - Aperture parameter of the Sobel derivative used.
- **k** - Harris detector free parameter in the equation.

```python
gray = cv2.cvtColor(img,cv.COLOR_BGR2GRAY) #Chuyn nh v nh thang xm

# Chuyn i hnh nh thnh cc gi tr floating-point v phng php cornerHarris s nhn loi gi tr ny. 
# Ch khng phi l gi tr trc tip khi cv2.cvtColor.
gray = np.float32(gray) 
dst = cv2.cornerHarris(gray,2,3,0.04) #c v2.cornerHarris(img_gray_float32, blockSize, ksize, k)
```

- S dng th vin c sn trong openCV cornerHarris  tm ra ma trn  sng 
tng ng vi mi pixel tun theo cng thc **R = det(M) - k(trace(M))$^2$**

  *  Nu |R| nh, iu ny xy ra khi $\lambda_{1}$ v $\lambda_{2}$ u nh ==> Vng mt phng
  *  Nu R < 0, iu ny xy ra khi $\lambda_{1}$ ln hn rt nhiu so vi $\lambda_{2}$ hoc ngc li ==> Vng cnh do ch thay i cng  theo 1 chiu
  *  Nu R ln,  iu ny xy ra khi $\lambda_{1}$ v $\lambda_{2}$ u ln v $\lambda_{1}$ ~ $\lambda_{2}$ ==> Vng gc do theo hng no th cng  cng thay i.

```python
# dilate  gin cc nh tm c ra  nh du cc gc(Khng quan trng).
dst = cv2.dilate(dst,None)
```

```python
# Vi mi gi tr pixel ln hn gi tr ln nht nhn vi h s threshold ( y l 0.01) ta s nh du bng pixel mu .
img[dst>0.01*dst.max()]=[0,0,255]
```

```python
cv2_imshow(img)
```

<h1 align='center'><b>Corner with SubPixel Accuracy<b></h1>

```python
filename = '/gdrive/MyDrive/Colab Notebooks/Computer Vision/Week 2/ChessBoard.png'
img = cv2.imread(filename)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
# Tm Harris corners
gray = np.float32(gray)
dst = cv2.cornerHarris(gray,2,3,0.04)
dst = cv2.dilate(dst,None)

# p dng nh phn nh threshold
ret, dst = cv2.threshold(dst,0.01*dst.max(),255,0)

# p kiu dst v dng unsigned int
dst = np.uint8(dst)
```

```python
# Tm cc centroids
ret, labels, stats, centroids = cv2.connectedComponentsWithStats(dst)
```

```python
# Xc nh cc tiu ch khi no dng lp. 
# Dng lp sau mt s ln lp c ch nh hoc t c  chnh xc nht nh
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.001)
corners = cv2.cornerSubPix(gray,np.float32(centroids),(5,5),(-1,-1),criteria)
```

```python
# Now draw them
res = np.hstack((centroids,corners)) # Ni 2 array theo trc x 
res = np.int0(res)
img[res[:,1],res[:,0]]=[0,0,255]
img[res[:,3],res[:,2]] = [0,255,0]
```

```python
cv2_imshow(img)
```
